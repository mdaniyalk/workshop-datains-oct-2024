### Loading the model and it's tokenizer in quantized setup!


```python
# setting up the config for 4-bit quantization of Qlora
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

# Configuring bitsandbytes for 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,  # Enable double quantization
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")

model_id = "microsoft/Phi-3.5-mini-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config, # Disable this if theres enough VRAM
    torch_dtype=torch.float16,
    trust_remote_code=True,
).to(device)

model
```

#### Prepare model for PEFT fine-tuning

```python
model.train()
```

```python
def print_trainable_parameters(model):
    """
    Prints the number of trainable parameters in the model.
    """
    trainable_params = 0
    all_param = 0
    for _, param in model.named_parameters():
        all_param += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    print(
        f"trainable params: {trainable_params} || all params: {all_param} || trainable%: {100 * trainable_params / all_param}"
    )
```

### Setup `LoraConfig`

```python
from peft import LoraConfig, get_peft_model

config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],  # parameters specific to llama
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)

model = get_peft_model(model, config)
print_trainable_parameters(model)
```

## Step 2) Fine-tuning process 💥


```python
# Load the dataset from HF
from datasets import load_dataset

data = load_dataset("timdettmers/openassistant-guanaco", split='train[:10]')  # Use only the first 10 samples
data = data.map(lambda samples: tokenizer(samples["text"]), batched=True)
```

```python
data
```

## Training

For the sake of the demo, we just ran it for 10 steps just to showcase how to use this integration with existing tools on the HF ecosystem.

```python
import transformers

tokenizer.pad_token = tokenizer.eos_token

trainer = transformers.Trainer(
    model=model,
    train_dataset=data,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=2,
        max_steps=10,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=1,
        output_dir="./results",  # change it to your desired repo!
        optim="paged_adamw_8bit",
    ),
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
)
model.config.use_cache = False  # silence the warnings. Please re-enable for inference!
trainer.train()
```

## Usage Example

```python
model.config.use_cache = True
model.eval();
```

```python
from transformers import GenerationConfig

max_new_tokens = 300
top_p = 0.9
temperature = 0.7
user_question = "¿Qué es GPT y transformadores?"

# Define the conversation template
prompt = (
    "A chat between a curious human and an artificial intelligence assistant. "
    "The assistant gives helpful, detailed, and polite answers to the user's questions. "
    "### Human: {user_question}\n"
    "### Assistant: "
)

def format_response(text):
    """
    Format the response to make it more readable by adding line breaks
    and cleaning up unnecessary symbols.
    """
    # Split into paragraphs based on periods, add line breaks after complete sentences
    formatted_text = text.replace("###", "").replace("Assistant:", "").strip()

    # Break the response into sentences to make it easier to read
    sentences = formatted_text.split(". ")

    # Join sentences into paragraphs
    formatted_response = "\n\n".join(sentences)

    return formatted_response

def generate(model, user_question, max_new_tokens=max_new_tokens, top_p=top_p, temperature=temperature):
    # Format the prompt with the user's question
    inputs = tokenizer(prompt.format(user_question=user_question), return_tensors="pt").to("cuda")

    # Generate response
    outputs = model.generate(
        **inputs,
        generation_config=GenerationConfig(
            do_sample=True,
            max_new_tokens=max_new_tokens,
            top_p=top_p,
            temperature=temperature,
        ),
    )

    # Decode the generated tokens into text
    text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Format the response for readability
    formatted_text = format_response(text)

    # Print the formatted response
    print(formatted_text)
    # return formatted_text

# Run the function
generate(model, user_question)
```
